```python
import requests
import json
import time

def chat(prompt, max_tokens=512, temp=0.7):
    url = "http://127.0.0.1:8000/chat"
    payload = {
        "prompt": prompt,
        "max_new_tokens": max_tokens,
        "temperature": temp
    }
    start = time.time()
    response = requests.post(url, json=payload, stream=True)
    text = ""
    for chunk in response.iter_content(chunk_size=None, decode_unicode=True):
        if chunk:
            text += chunk
            print(chunk, end="", flush=True)
    elapsed = time.time() - start
    tokens = len(text.split())  # rough estimate
    print(f"\n\n~{tokens / elapsed:.1f} tokens/s")
    return text
```

```python
chat("Write a clean .NET 8 minimal API that returns JSON with current UTC time and server hostname.")
```
